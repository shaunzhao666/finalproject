```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
```

```python
hd_df = pd.read_csv("https://raw.githubusercontent.com/shaunzhao666/midterm_object/code/heart.csv")
hd_df.describe()
```

```python
hd_df.head()
```

```python
de_val = ['age', 'exang', 'thal', 'restecg', 'cp', 'slope', 'ca']
con_val = ['trestbps', 'chol', 'thalach', 'oldpeak']
```

```python
hd_pre = hd_df.drop(columns="fbs")
hd_pre
```

```python
dic = []
for i in con_val:
    a = np.array(hd_pre.loc[:, i]).reshape(hd_df.shape[0], 1)
    hd_pre.loc[:, i] = StandardScaler().fit_transform(a)
    pre = StandardScaler().fit(a)
    dic.append((i, pre))
for j in de_val:
    a = np.array(hd_pre.loc[:, j]).reshape(hd_df.shape[0], 1)
    hd_pre.loc[:, j] = MinMaxScaler().fit_transform(a)
    pre = MinMaxScaler().fit(a)
    dic.append((j, pre))
hd_pre
```

```python
name_cla = [
    "Nearest Neighbors",
    "linear SVM",
    "RBF SVM",
    "Neural Network",
    "Naive Bayes",
    "Decision Tree",
    "logistic regression"
    ]


classifiers = [
    KNeighborsClassifier(2),
    SVC(kernel='linear', C=0.025),
    SVC(gamma=0.001),
    MLPClassifier(max_iter=1000),
    GaussianNB(),
    DecisionTreeClassifier(max_depth=3),
    LogisticRegression
]
```

```python
Y_df = hd_pre["target"].values.copy()
X_df = hd_pre.iloc[:, 0:12].values.copy()
```

```python
# KNN
## test size
test_size = [0.05, 0.1, 0.3, 0.5, 0.7, 0.9]
ele = np.zeros((len(test_size), 5))
n=0
for i in test_size:
    X_train, X_test, Y_train, Y_test = train_test_split(X_df, Y_df, test_size=i, random_state=1)
    classify = KNeighborsClassifier(2).fit(X_train, Y_train)
    prediction = classify.predict(X_test)
    ele[n, 0] = i
    ele[n, 1] = classify.score(X_test, Y_test)
    ele[n, 2] = metrics.precision_score(Y_test, prediction)
    ele[n, 3] = metrics.recall_score(Y_test, prediction)
    ele[n, 4] = metrics.f1_score(Y_test, prediction)
    n+=1
ele_df = pd.DataFrame(ele, columns=["test_size", "accuracy", "precision", "recall", "f1_score"])
ele_df.to_csv("knn_elements.csv")

ele = pd.read_csv("knn_elements.csv")
ele_df
```

```python
test_size = [0.05, 0.1, 0.3, 0.5, 0.7, 0.9]
c = [1, 10, 100]
ele = np.zeros((len(c)*len(test_size), 6))
n = 0
for i in test_size:
    for j in c:
        X_train, X_test, Y_train, Y_test = train_test_split(X_df, Y_df, test_size=i, random_state=1)
        classify = SVC(kernel='linear', C=j).fit(X_train, Y_train)
        prediction = classify.predict(X_test)
        ele[n, 0] = i
        ele[n, 1] = j
        ele[n, 2]= classify.score(X_test, Y_test)
        ele[n, 3] = metrics.precision_score(Y_test, prediction)
        ele[n, 4] = metrics.recall_score(Y_test, prediction)
        ele[n, 5] = metrics.f1_score(Y_test, prediction)
        n += 1
ele_df = pd.DataFrame(ele, columns=["test_size", "C", "accuracy", "precision", "recall", "f1_score"])
ele_df.to_csv("lsvm_element.csv")
ele_df
```

```python
# svm
test_size = [0.05, 0.1, 0.3, 0.5, 0.7, 0.9]
gamma_svm = [0.001, 0.02, 0.03, 0.04]
c_svm = [1, 10, 50, 100]
ele = np.zeros((len(c_svm)*len(gamma_svm)*len(test_size), 7))
n = 0
for i in test_size:
    for j in gamma_svm:
        for z in c_svm: 
            X_train, X_test, Y_train, Y_test = train_test_split(X_df, Y_df, test_size=i, random_state=1)
            classify = SVC(gamma=j, C=z).fit(X_train, Y_train)
            prediction = classify.predict(X_test)
            ele[n, 0] = i
            ele[n, 1] = j
            ele[n, 2] = z
            ele[n, 3]= classify.score(X_test, Y_test)
            ele[n, 4] = metrics.precision_score(Y_test, prediction)
            ele[n, 5] = metrics.recall_score(Y_test, prediction)
            ele[n, 6] = metrics.f1_score(Y_test, prediction)
            n += 1
ele_df = pd.DataFrame(ele, columns=["test_size", "gamma", "C", "accuracy", "precision", "recall", "f1_score"])
ele_df.to_csv("svm_element.csv")
```

```python
# nn
test_size = [0.05, 0.1, 0.3, 0.5, 0.7, 0.9]
laysize = [10, 50, 100, 200]
max_iter = [200, 500, 1000]

ele = np.zeros((len(laysize)*len(test_size)*len(max_iter), 7))
n = 0
for i in test_size:
    for j in laysize: 
        for z in max_iter:
            X_train, X_test, Y_train, Y_test = train_test_split(X_df, Y_df, test_size=i, random_state=1)
            classify = MLPClassifier(hidden_layer_sizes=(j, ), max_iter=z).fit(X_train,  Y_train)
            prediction = classify.predict(X_test)
            ele[n, 0] = i
            ele[n, 1] = j
            ele[n, 2] = z
            ele[n, 3]= classify.score(X_test, Y_test)
            ele[n, 4] = metrics.precision_score(Y_test, prediction)
            ele[n, 5] = metrics.recall_score(Y_test, prediction)
            ele[n, 6] = metrics.f1_score(Y_test, prediction)
            n += 1
ele_df = pd.DataFrame(ele, columns=["test_size", "hidden_layers", "max_iteration",  "accuracy", "precision", "recall", "f1_score"])
ele_df.to_csv("nn_element.csv")
ele_df
```
